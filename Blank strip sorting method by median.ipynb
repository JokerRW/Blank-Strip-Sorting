```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import ExcelWriter
#Read file
fileName = 'BD7618A221071_1598_Job1'# 請在這裡更換檔名
filepath0='C:\\Users\\richard.weng\\Documents\\Python Scripts\\python_projects\\(2) Blank strip\\'+fileName+'.xls'
df=pd.read_excel(filepath0, encoding = 'utf-8')
#Remove first 4 rows
df = df.iloc[4:,:]
#Change column_name 
column_name = df[df['Date'] == 'Index']
df.columns = column_name.iloc[0]
#Delete first row
df = df.iloc[1:,:]

#Design spec for strip selection
sigma = 3 # 請在這裡調整強度
median = df['Measurement Results (mm2)'].median()
Upper = median+median*0.012*sigma
Lower = median-median*0.012*sigma
print('中位數:',median)
print('上限:',Upper)
print('下限:',Lower)
print('-------------','\n')
#Filter_ok strip
df1 = df[df['Measurement Results (mm2)']<=Upper]
df_want = df1[df1['Measurement Results (mm2)']>=Lower]
#Filter NG strip
df3 = df[df['Measurement Results (mm2)']>Upper]
df4 = df[df['Measurement Results (mm2)']<Lower]
df_dele = pd.concat([df3,df4])

#Check Original CV
STD = df['Measurement Results (mm2)'].std()
Mean = df['Measurement Results (mm2)'].mean()
CV = STD/Mean*100
print('原STD:',STD)
print ('原Mean:',Mean)
print ('原CV:',CV)
print('-------------','\n')

#Check New CV
STD = df_want['Measurement Results (mm2)'].std()
Mean = df_want['Measurement Results (mm2)'].mean()
CV = STD/Mean*100
print('新STD:',STD)
print ('新Mean:',Mean)
print ('新CV:',CV)
print('-------------','\n')
print('總共刪了:', '%d'%(len(df_dele)),'片')

#Save the new result into new Excel
writer = ExcelWriter(fileName+'_new'+'.xlsx')
df_want.to_excel(writer,'OK_Strip',index=False)
df_dele.to_excel(writer,'NG_Strip',index=False)

#plot
workbook=writer.book
worksheet= writer.sheets['OK_Strip']
chart=workbook.add_chart({'type':'scatter'})
chart.add_series({'name':'=OK_Strip!$E$1','categories':'=OK_Strip!$A$2:$A$313','values':'=OK_Strip!$E$2:$E$313'})
chart.set_title ({'name': 'Blank Strip AOI Scatter'})
chart.set_x_axis({'name':'Index'})
chart.set_y_axis({'name':'AOI'})
chart.set_legend({'position':'top'})
worksheet.insert_chart('J3:T17',chart)
writer.save()                               
```
